```python
en = LabelEncoder()

col_train =  ["term", "home ownership", "loan status", "purpose"]
col_test =  ["term", "home ownership", "purpose"]

for cols in col_train:
    df_train[cols] = en.fit_transform(df_train[cols])


for cols in col_test:
    df_test[cols] = en.fit_transform(df_test[cols])
```

```python
imputer = KNNImputer(n_neighbors= 3)
df_train_clean = pd.DataFrame(imputer.fit_transform(df_train), columns = df_train.columns)
df_train_clean.isnull().sum()
```

```python
imputer = KNNImputer(n_neighbors= 3)
df_test_clean = pd.DataFrame(imputer.fit_transform(df_test), columns = df_test.columns)
df_test_clean.isnull().sum()
```

```python
X = df_train_clean.drop("loan status", axis= 1)
y_train = df_train_clean["loan status"]
```

```python
sc = StandardScaler()
```

```python
X_test = df_test_clean
```

```python
X_train = sc.fit_transform(X)
X_test = sc.transform(X_test)
```

```python
rfc = RandomForestClassifier(criterion = "entropy", random_state = 75)
```

```python
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
param_grid = {"n_estimators": [50, 100, 150, 200],
              "max_depth": [None, 10, 20, 30],
              "min_samples_split": [2, 5, 10],
              "min_samples_leaf": [1, 2, 4],
              "max_features": ["sqrt", "log2"],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]
              }
```

```python
Grid_Search = GridSearchCV(estimator= rfc, param_grid= param_grid, cv= 2, scoring= "accuracy")
Grid_Search.fit(X_train, y_train)
```

```python
best_rf_model = Grid_Search.best_estimator_
best_rf_model.fit(X_train, y_train)
```

```python
y_pred = best_rf_model.predict(X_test)
```
