**Renaming by passing the new values**

**Replace a default string in the header**

```python
df_3 = df
df_3.columns = ["$Week", "&eLearn", "$DS", "$ML", "#AI", "@DL"]
df_3.head(2)
```

```python
df_3.columns = df_3.columns.str.replace("$","")
df_3.head(2)
```

```python
df_3.columns = df_3.columns.str.replace("&","_")
df_3.head(2)
```
