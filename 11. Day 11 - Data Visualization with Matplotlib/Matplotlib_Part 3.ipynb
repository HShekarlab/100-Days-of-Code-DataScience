```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([11, 36, 9, 43, 15, 84, 66, 7, 23])
y = np.array([52, 74, 31, 12, 29, 46, 13, 9, 68])

plt.scatter(x, y)

plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
x = np.array([11, 36, 9, 43, 15, 84, 66, 7, 23])
y = np.array([52, 74, 31, 12, 29, 46, 13, 9, 68])
colors = np.array(["b", "r", "y", "c", "m", "k", "g", "silver", "orange"])

plt.scatter(x, y, c= colors)

plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
x = np.array([11, 36, 9, 43, 15, 84, 66, 7, 23])
y = np.array([52, 74, 31, 12, 29, 46, 13, 9, 68])
colors = np.array(["b", "r", "y", "c", "m", "k", "g", "silver", "orange"])

plt.scatter(x, y, c= colors, cmap= "viridis")
plt.colorbar()

plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
x = np.array([11, 36, 9, 43, 15, 84, 66, 7, 23])
y = np.array([52, 74, 31, 12, 29, 46, 13, 9, 68])
size = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50])

plt.scatter(x, y, s= size)

plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
x = np.array([11, 36, 9, 43, 15, 84, 66, 7, 23])
y = np.array([52, 74, 31, 12, 29, 46, 13, 9, 68])
size = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50])

plt.scatter(x, y, s= size, alpha= 0.8)

plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
x = np.random.randint(100, size= (100))
y = np.random.randint(100, size= (100))
colors = np.random.randint(100, size= (100))
sizes = 10 * np.random.randint(100, size= (100))

plt.scatter(x, y, c= colors, s= sizes, alpha= 0.5, cmap= "spring")
plt.colorbar()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([52, 74, 31, 12])

plt.bar(x, y)

plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([52, 74, 31, 12])

plt.bar(x, y, color= "b")

plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([52, 74, 31, 12])

plt.barh(x, y, color= "m")

plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([52, 74, 31, 12])

plt.bar(x, y, color= "r", width= 0.2)

plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
x = np.random.normal(150, 15, 230)

plt.hist(x)
```

```python
x = np.array([35, 25, 25, 15])
fruits = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(x, labels = fruits, startangle = 90)
```

```python
x = np.array([35, 25, 25, 15])
fruits = ["Apples", "Bananas", "Cherries", "Dates"]
explod = [0.2, 0, 0, 0]

plt.pie(x, labels = fruits, explode = explod , shadow= True)
```

```python
x = np.array([35, 25, 25, 15])
fruits = ["Apples", "Bananas", "Cherries", "Dates"]
color = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(x, labels = fruits, colors = color)
plt.legend(title="Four Fruits")
```
