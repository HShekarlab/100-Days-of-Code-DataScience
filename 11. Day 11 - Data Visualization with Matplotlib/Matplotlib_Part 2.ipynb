```python
x = np.array(np.arange(1, 100))

y_1 = x ** 2
y_2 = x ** 3

plt.subplot(1, 2, 1)
plt.plot(x, y_1, "b*")

plt.subplot(1, 2, 2)
plt.plot(y_2, x, "c+")
```

```python
x = np.array(np.arange(1, 100))

y_1 = x ** 2
y_2 = x ** 3

plt.subplot(2, 1, 1)
plt.plot(x, y_1, "b.")

plt.subplot(2, 1, 2)
plt.plot(y_2, x, "g-")
```
